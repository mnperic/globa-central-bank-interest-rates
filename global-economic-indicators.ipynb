```python
# Import dependencies and setup

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import matplotlib.dates as matplotlibdates
from matplotlib.dates import date2num
```

```python
## Read .csv files and store as DataFrames
```

```python
# banking sector leverage
# business confidence index (BCI)
# composite leading indicator (CLI)
# consumer confidence index (CCI)
# exchange rates
# financial corporate debt to equity ratio
# general government deficit
# gross domestic product (GDP)
# quarterly gross domestic product (QGDP)
# household debt
# household disposable income
# inflation (CPI)
# long term interest rates
# part-time employment rate
# short term interest rates
# unemployment rate
# youth_unemployment rate

banking_sector_leverage_path = "Resources/banking_sector_leverage.csv"
bsl_df = pd.read_csv(banking_sector_leverage_path)

business_confidence_index_path = "Resources/business_confidence_index_bci.csv"
bci_df = pd.read_csv(business_confidence_index_path)

composite_leading_indicator_path = "Resources/composite_leading_indicator_cli.csv"
cli_df = pd.read_csv(composite_leading_indicator_path)

consumer_confidence_index_path = "Resources/consumer_confidence_index_cci.csv"
cci_df = pd.read_csv(consumer_confidence_index_path)

exchange_rates_path = "Resources/exchange_rates.csv"
exchange_rates_df = pd.read_csv(exchange_rates_path)

financial_corporate_debt_to_equity_path = "Resources/financial_corporate_debt_to_equity.csv"
fcdter_df = pd.read_csv(financial_corporate_debt_to_equity_path)

general_government_deficit_path = "Resources/general_government_deficit.csv"
ggd_df = pd.read_csv(general_government_deficit_path)

gdp_path = "Resources/gross_domestic_product_gdp.csv"
gdp_df = pd.read_csv(gdp_path)

quarterly_gdp_path = "Resources/quarterly_gross_domestic_product_gdp.csv"
qgdp_df = pd.read_csv(quarterly_gdp_path)

household_debt_path = "Resources/household_debt.csv"
household_debt_df = pd.read_csv(household_debt_path)

household_disposable_income_path = "Resources/household_disposable_income.csv"
hdi_df = pd.read_csv(household_disposable_income_path)

inflation_cpi_path = "Resources/inflation.csv"
cpi_df = pd.read_csv(inflation_cpi_path)

long_term_interest_rates_path = "Resources/long_term_interest_rates.csv"
ltir_df = pd.read_csv(long_term_interest_rates_path)

part_time_employment_rate_path = "Resources/part-time_employment_rate.csv"
pter_df = pd.read_csv(part_time_employment_rate_path)

short_term_interest_rates_path = "Resources/short_term_interest_rates.csv"
stir_df = pd.read_csv(short_term_interest_rates_path)

unemployment_rate_path = "Resources/unemployment_rate.csv"
unemployment_df = pd.read_csv(unemployment_rate_path)

youth_unemployment_rate_path = "Resources/youth_unemployment_rate.csv"
youth_unemployment_df = pd.read_csv(youth_unemployment_rate_path)
```

```python
# # Convert all 'TIME' columns datatype to datetime

bsl_df['TIME'] = pd.to_datetime(bsl_df['TIME'], format='%Y')
bci_df['TIME'] = pd.to_datetime(bci_df['TIME'], format='%Y-%m')
cli_df['TIME'] = pd.to_datetime(cli_df['TIME'], format='%Y-%m')
cci_df['TIME'] = pd.to_datetime(cci_df['TIME'], format='%Y-%m')
exchange_rates_df['TIME'] = pd.to_datetime(exchange_rates_df['TIME'], format='%Y')
fcdter_df['TIME'] = pd.to_datetime(fcdter_df['TIME'], format='%Y')
ggd_df['TIME'] = pd.to_datetime(ggd_df['TIME'], format='%Y')
gdp_df['TIME'] = pd.to_datetime(gdp_df['TIME'], format='%Y')
qgdp_df['TIME'] = pd.to_datetime(qgdp_df['TIME'])
household_debt_df['TIME'] = pd.to_datetime(household_debt_df['TIME'], format='%Y')
hdi_df['TIME'] = pd.to_datetime(hdi_df['TIME'], format='%Y')
cpi_df['TIME'] = pd.to_datetime(cpi_df['TIME'], format='%Y')
ltir_df['TIME'] = pd.to_datetime(ltir_df['TIME'], format='%Y')
pter_df['TIME'] = pd.to_datetime(pter_df['TIME'], format='%Y')
stir_df['TIME'] = pd.to_datetime(stir_df['TIME'], format='%Y')
unemployment_df['TIME'] = pd.to_datetime(unemployment_df['TIME'], format='%Y')
youth_unemployment_df['TIME'] = pd.to_datetime(youth_unemployment_df['TIME'], format='%Y')
```

```python
# Create new DataFrames based on locations (countries)

# bsl_df['LOCATION'].unique()
    # Countries = [AUS, CAN, DEU, FRA, GBR, ITA, JPN, USA]
    
aus_bsl_df = bsl_df[bsl_df.LOCATION == 'AUS'] #Australia
can_bsl_df = bsl_df[bsl_df.LOCATION == 'CAN'] #Canada
deu_bsl_df = bsl_df[bsl_df.LOCATION == 'DEU'] #Germany
fra_bsl_df = bsl_df[bsl_df.LOCATION == 'FRA'] #France
gbr_bsl_df = bsl_df[bsl_df.LOCATION == 'GBR'] #Great Britain
ita_bsl_df = bsl_df[bsl_df.LOCATION == 'ITA'] #Italy
jpn_bsl_df = bsl_df[bsl_df.LOCATION == 'JPN'] #Japan
usa_bsl_df = bsl_df[bsl_df.LOCATION == 'USA'] #USA
```

```python
# Plotting – BSL (Australia)

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.plot(aus_bsl_df['TIME'], aus_bsl_df['VALUE'], color='blue', linewidth=3)

# Labels
plt.title("Banking Sector Leverage (Equity Multiplier Ratio) – Australia, yearly average (1995–2020)", fontdict = font_1)
plt.xlabel("Year", fontdict = font_2)
plt.ylabel("% of net value added", fontdict = font_2)

# Highlights
plt.axvspan(date2num(datetime(2007,1,1)), date2num(datetime(2009,1,1)), 
           label="2008 Global Financial Crisis", color="red", alpha=0.1)

plt.axvspan(date2num(datetime(2012,1,1)), date2num(datetime(2015,1,1)), 
           label="2012–2014 European Debt Crisis", color="orange", alpha=0.1)
plt.legend()

# Aesthetics
matplotlib.pyplot.grid(visible=True, which='both', axis='both', color='blue', alpha=0.1, linestyle='--', linewidth=1)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Save and show
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/global-economic-indicators/Images/bsl_australia.png", dpi=300)
plt.show()
```
