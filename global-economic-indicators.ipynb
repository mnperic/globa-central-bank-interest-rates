```python
# Convert all 'TIME' columns datatype to datetime

bsl_df['TIME'] = pd.to_datetime(bsl_df['TIME'], format='%Y')
```

```python
## Banking Sector Leverage
```

```python
# Create new DataFrames based on locations (countries)

# bsl_df['LOCATION'].unique()
    # Countries = AUS, CAN, DEU, FRA, GBR, ITA, JPN, USA
    
aus_bsl_df = bsl_df[bsl_df.LOCATION == 'AUS'] #Australia
can_bsl_df = bsl_df[bsl_df.LOCATION == 'CAN'] #Canada
deu_bsl_df = bsl_df[bsl_df.LOCATION == 'DEU'] #Germany
fra_bsl_df = bsl_df[bsl_df.LOCATION == 'FRA'] #France
gbr_bsl_df = bsl_df[bsl_df.LOCATION == 'GBR'] #Great Britain
ita_bsl_df = bsl_df[bsl_df.LOCATION == 'ITA'] #Italy
jpn_bsl_df = bsl_df[bsl_df.LOCATION == 'JPN'] #Japan
usa_bsl_df = bsl_df[bsl_df.LOCATION == 'USA'] #USA
```

```python
# Plotting

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.plot(aus_bsl_df['TIME'], aus_bsl_df['VALUE'], color='blue', linewidth=3)

# Labels
plt.title("Banking Sector Leverage/Equity Multiplier Ratio (Australia), 1995–2020 (yearly average)", fontdict = font_1)
plt.xlabel("Year", fontdict = font_2)
plt.ylabel("% net value added", fontdict = font_2)

# Highlights
plt.axvspan(date2num(datetime(1997,1,1)), date2num(datetime(2000,1,1)), 
           label="Dot-Com Bubble", color="red", alpha=0.1)
plt.axvspan(date2num(datetime(2007,1,1)), date2num(datetime(2009,1,1)), 
           label="Global Financial Crisis", color="orange", alpha=0.1)
plt.legend()

# Aesthetics
matplotlib.pyplot.grid(visible=True, which='both', axis='both', 
                       color='blue', alpha=0.1, dash_capstyle='butt',
                       linestyle='--', linewidth=1)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()



# Save and show
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/global-economic-indicators/Images/bsl_aus.png", dpi=300)
plt.show()
```
